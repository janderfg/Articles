# Convertendo coluna

## Dados contínuos para dados categóricos

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Nome': ['Lucas', 'Pedro', 'Maria', 'Paulo', 'Eliana', \
                            'Mateus', 'Jander', 'Lucas', 'Dayana', 'Ricardo'],
                   'Idade': [16, 19, 23, 17, 32, 65, 36, 62, 33, 18]})
```

```python
df
```

```python
df['Faixa'] = pd.cut(df['Idade'],
                     bins=[0, 17, 59, 130],
                     labels=['Criança', 'Adulto', 'Idoso'])
```
