Em determinadas situações é necessário efetuar separação de dados de 1 coluna para 2 colunas, como por exemplo "Nome" e "Sobrenome".


Segue abaixo um exemplo prático e rápido para resolver este tipo de situação:

```python
# Importando a biblioteca necessária
import pandas as pd
```

```python
# Criando um dataset de exemplo
df = pd.DataFrame({'Nome': ['Dayana Guedes', 'Camila Fernandes', 'Felipe Luiz Guimarães', 'Dayana Lima', \
                            'Jander Guimarães', 'Paulo de Souza', 'Suzielaine de Oliveira Guimarães', \
                            'Marcos Paulo Filho', 'Oswaldo Neto', 'Gabriela Martins'],})
```

```python
# Visualizando o dataset criado
df
```

```python
# Efetuando a separação com "Nome" e "Sobrenome"
df['Primeiro Nome'] = df['Nome'].str.split(n=1).str[0]
df['Sobrenome'] = df['Nome'].str.split(n=1).str[1]
```

```python
# Resultado da separação
df
```
